19.Write a NumPy program to merge three given NumPy arrays of same shape

```python
import numpy as np

# Create three NumPy arrays of the same shape
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.array([[7, 8, 9],
                   [10, 11, 12]])

array3 = np.array([[13, 14, 15],
                   [16, 17, 18]])

# Merge the three arrays along the first axis (axis=0)
merged_array = np.concatenate((array1, array2, array3), axis=0)

# Print the merged array
print("Merged Array:")
print(merged_array)
```

20.Write a NumPy program to combine last element with first element of two given ndarray with different shapes.

```python
import numpy as np

# Create two NumPy arrays with different shapes
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6, 7])

# Combine last element of array1 with the first element of array2
combined_array = np.concatenate((array1[-1:], array2[:1]))

# Print the combined array
print("Combined Array:")
print(combined_array)
```

3.Write a Pandas program to get the powers of an array values element-wise. Note: First array elements raised to powers from second array Sample data: {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}

```python
import pandas as pd
data = {'X': [78, 85, 96, 80, 86],
        'Y': [84, 94, 89, 83, 86],
        'Z': [86, 97, 96, 72, 83]}
df = pd.DataFrame(data)
powers_array = [2, 3, 2]
result = df.pow(powers_array)
print("DataFrame with Element-wise Powers:")
print(result)
```

13.Apply and explore various plotting functions on Pima Indians Diabetes data set for performing the following: a) Correlation and scatter plots b) Histograms c) Three-dimensional plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load the Pima Indians Diabetes dataset
diabetes_data_url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
diabetes_df = pd.read_csv(diabetes_data_url, header=None, names=column_names)

# Display the first few rows of the dataset
print("First few rows of the Pima Indians Diabetes dataset:")
print(diabetes_df.head())

# Set style for Seaborn plots
sns.set(style="whitegrid")

# a) Correlation and Scatter Plots
correlation_matrix = diabetes_df.corr()

plt.figure(figsize=(12, 8))

# Heatmap for correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Correlation Matrix Heatmap")
plt.show()

# Scatter Plot Matrix
sns.pairplot(diabetes_df, hue='Outcome', markers=["o", "s"], palette="husl", diag_kind="kde")
plt.suptitle("Scatter Plot Matrix", y=1.02)
plt.show()

# b) Histograms
plt.figure(figsize=(12, 8))
diabetes_df.hist(bins=20, edgecolor='black', linewidth=1.2, grid=False)
plt.suptitle("Histograms of Features", y=1.02)
plt.tight_layout()
plt.show()

# c) Three-dimensional Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(diabetes_df['BMI'], diabetes_df['Age'], diabetes_df['Glucose'], c=diabetes_df['Outcome'], cmap='viridis')
ax.set_xlabel('BMI')
ax.set_ylabel('Age')
ax.set_zlabel('Glucose')
ax.set_title("Three-dimensional Plot of BMI, Age, and Glucose")
plt.show()
```

8.Use the diabetes data set from UCI data set for performing the following:
Apply Univariate analysis:  Frequency
 Mean,  Median,  Mode,  Variance  Standard Deviation
 Skewness and Kurtosis

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Load the diabetes dataset
diabetes_data = pd.read_csv("/content/drive/MyDrive/diabetes.csv")

# Univariate Analysis
column_of_interest = 'Glucose'  # Replace with the column you want to analyze

# Frequency
frequency = diabetes_data[column_of_interest].value_counts()

# Mean, Median, Mode
mean_value = diabetes_data[column_of_interest].mean()
median_value = diabetes_data[column_of_interest].median()
mode_value = diabetes_data[column_of_interest].mode()

# Variance and Standard Deviation
variance_value = diabetes_data[column_of_interest].var()
std_deviation_value = diabetes_data[column_of_interest].std()

# Skewness and Kurtosis
skewness_value = skew(diabetes_data[column_of_interest])
kurtosis_value = kurtosis(diabetes_data[column_of_interest])

# Display results
print(f"Frequency:\n{frequency}\n")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value.iloc[0] if not mode_value.empty else 'No mode'}")
print(f"Variance: {variance_value:.2f}")
print(f"Standard Deviation: {std_deviation_value:.2f}")
print(f"Skewness: {skewness_value:.2f}")
print(f"Kurtosis: {kurtosis_value:.2f}")
```
